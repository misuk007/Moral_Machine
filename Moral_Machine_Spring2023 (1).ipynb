```python
import random

class MMGenerator:
    MAN = 'a'
    WOMAN = 'b'
    BOY = 'c'
    GIRL = 'd'
    ELDERLY_MAN = 'e'
    ELDERLY_WOMAN = 'f'
    OBESE_MAN = 'g'
    OBESE_WOMAN = 'h'
    MALE_EXECUTIVE = 'i'
    FEMALE_EXECUTIVE = 'j'
    MALE_DOCTOR = 'k'
    FEMALE_DOCTOR = 'l'
    MALE_JOGGER = 'm'
    FEMALE_JOGGER = 'n'
    PREGNANT_WOMAN = 'o'
    HOMELESS_PERSON = 'p'
    CRIMINAL = 'q'
    BABY = 'r'
    DOG = 's'
    CAT = 't'

    def getScenario(self):
        scenario = random.randint(0,2)
        legal = random.randint(0,2)
        return scenario + legal * 10

    def getGroup(self):
        group_size = random.randint(1,5)
        group = ''
        for i in range(group_size):
            rand_choice = random.randint(0,20)
            if rand_choice == 0:
                group += MMGenerator.MAN
            elif rand_choice == 1:
                group += MMGenerator.WOMAN
            elif rand_choice == 2:
                group += MMGenerator.BOY
            elif rand_choice == 3:
                group += MMGenerator.GIRL
            elif rand_choice == 4:
                group += MMGenerator.ELDERLY_MAN
            elif rand_choice == 5:
                group += MMGenerator.ELDERLY_WOMAN
            elif rand_choice == 6:
                group += MMGenerator.OBESE_MAN
            elif rand_choice == 7:
                group += MMGenerator.OBESE_WOMAN
            elif rand_choice == 8:
                group += MMGenerator.MALE_EXECUTIVE
            elif rand_choice == 9:
                group += MMGenerator.FEMALE_EXECUTIVE
            elif rand_choice == 10:
                group += MMGenerator.MALE_DOCTOR
            elif rand_choice == 11:
                group += MMGenerator.FEMALE_DOCTOR
            elif rand_choice == 12:
                group += MMGenerator.MALE_JOGGER
            elif rand_choice == 13:
                group += MMGenerator.FEMALE_JOGGER
            elif rand_choice == 14:
                group += MMGenerator.PREGNANT_WOMAN
            elif rand_choice == 15:
                group += MMGenerator.HOMELESS_PERSON
            elif rand_choice == 16:
                group += MMGenerator.CRIMINAL
            elif rand_choice == 17:
                group += MMGenerator.BABY
            elif rand_choice == 18:
                group += MMGenerator.DOG
            elif rand_choice == 19:
                group += MMGenerator.CAT
        return group


class MoralMachine:

    def main(self):
        mm_generator = MMGenerator()
        scenario = mm_generator.getScenario()


        print("Scenario: ", scenario)
        g1 = mm_generator.getGroup()
        g2 = mm_generator.getGroup()

        print("Group 1: ", g1)
        print("Group 2: ", g2)

        final_decision =self.decideSwerve(scenario, g1, g2)

        if final_decision:
            print("Decision: Swerve to save more lives")
        else:
            print("Decision: Stay in the current lane.")

    def decideSwerve(self, scenario, g1, g2):
        if self.savingMoreLives(scenario, g1, g2):
            return True
        else:
            return False

    def savingMoreLives(self,scenario, g1, g2):
        result = False

        if scenario < 11:
            result = self.groupSize(g1,g2)
        elif scenario < 22:
            if 'o' in g1 or 'o' in g2:
                result = False
            else:
                result = self.groupSize(g1,g2)
        else:
            if 'q' in g1 or 'q' in g2:
                result = True
            else:
                result = self.groupSize(g1,g2)

        return result

    def groupSize(self,g1,g2):
        if len(g1) > len(g2):
            return True
        elif len(g1) < len(g2):
            return False
        else:
            return True

mm = MoralMachine()
mm.main()




















```
